#### Bullet 3: Explain why the back-propogation algorithm works on ms-cnn

### Problem b.3 - Explanation

In back-propagation, the gradient is calculated for each layer from the output layer to the input layer using chain rule. The CNNs have multiple layers that are related to each other. This construct of CNN is similar to the one that we have in MLPs. The variables of one layer depend on the variables of previous layer and chain rule can be applied. It is because of this similarity and relationships between multiple layers of CNN, Theano can calculate gradients of the CNN via backpropagation algorithm.

### Problem 2 - Comparing CNN and MLP

```python
from hw3b import test_lenet
from hw3a import test_mlp

nkerns = [16, 512]
filter_size = 3
cnn_comparison_output = test_lenet(learning_rate=0.05, n_epochs=1000, nkerns=nkerns,
            batch_size=200, verbose=True,filter_size=filter_size)

layer1_neurons = (32-filter_size+1)//2
layer2_neurons = (layer1_neurons-filter_size+1)//2
hidden_units = [nkerns[0]*layer1_neurons*layer1_neurons, nkerns[1]*layer2_neurons*layer2_neurons, 500]
hidden_cnt = 3

mlp_comparison_output = test_mlp(learning_rate=0.05, L1_reg=0.00, L2_reg=0.0001, n_epochs=200, 
                                 batch_size=128, n_hidden=hidden_units, n_hiddenLayers=hidden_cnt,
                                 verbose=True, smaller_set=True)
```
