### Bullet 2: Comparison between _tanh_ and _softmax_

```python
# TODO
from hw2b import load_data, test_mlp, LogisticRegression, HiddenLayer, MLP, myMLP
import theano.tensor as T
import gc
activationList = [T.tanh, T.nnet.softmax]
question2Output = []
for val in activationList:
    print val
    question2Output.append(test_mlp(learning_rate=0.05, L1_reg=0.00, L2_reg=0.0001, 
         n_epochs=200,batch_size=100, n_hidden=300, verbose=True,layer_cnt=2,activation=val))
print question2Output    
gc.collect()
```

### Nitesh Chauhan
As seen from the output below, test accuracy of tanh function is 80.965384615384608 and that of softmax function is 19.584615384615375. Hence, tanh function provides better convergence and test accuracy as compared to softmax function.

Also, I tried experimenting with different parameter values. 

With n_epochs=1000, learning_rate=0.01 and n_hidden=500, batch_size=100 it took a lot of time to execute.

With n_epochs=150, learning_rate=0.08, batch_size=100 and n_hidden=200, reduced the execution time but resulted in the configuration not converging in given epochs.

Then I tried with the configuration of n_epochs=200, batch_size=100, learning_rate=0.05 and n_hidden=300. It gave reasonable results with the model converging quickly. 

Higher epochs like 500 in this configuration resulted in infinitesimal improvement in the test accuracy.

### Bullet 4: Experiment with the number of hidden layers, but fix the total number of neurons

```python
# TODO
%matplotlib inline
import matplotlib.pyplot as plt
from hw2b import load_data, test_mlp, LogisticRegression, HiddenLayer, MLP, myMLP
import theano.tensor as T
import gc
gc.collect()

hidden_cnt = 1200
output = []
for val in range(8):
    print "Iteration",val
    hcount = hidden_cnt/(val+1)
    print "Hidden Units", hcount
    output.append(test_mlp(learning_rate=0.05, L1_reg=0.00, L2_reg=0.0001, 
         n_epochs=200,batch_size=100, n_hidden=hidden_cnt, verbose=True,layer_cnt=(val+1),activation=T.tanh))
    print "Test Accuracy",output
    gc.collect()

plt.plot([3,4,5,6,7,8,9,10], output, 'bo-')
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Test Accuracy,[%]')
plt.title('Effect of Number of Layers Total Number of Parameters in Hidden Layers is fixed to 1.2K')
plt.show()
```

### Bullet 5: Experiment with the number of neurons in hidden layers, but fix the number of hidden layers

```python
# TODO
%matplotlib inline
import matplotlib.pyplot as plt
from hw2b import load_data, test_mlp, LogisticRegression, HiddenLayer, MLP, myMLP
import theano.tensor as T
import gc

gc.collect()

question5Output = {}
for i in range(1,3,1):
    question5Output[i] = []
    for j in range(200,1200,200):
        print 'Number of Hidden Units-->',j
        temp = test_mlp(learning_rate=0.05, L1_reg=0.00, L2_reg=0.0001, 
         n_epochs=200,batch_size=100, n_hidden=j, verbose=True,layer_cnt=i,activation=T.tanh)
        question5Output[i].append(temp)
        print temp
        print question5Output
    
    plt.plot(range(200,1200,200), question5Output[i], 'bo-')
    plt.xlabel('Number of Neurons in the Hidden Layer')
    plt.ylabel('Test Accuracy,[%]')
    plt.title('Effect of Number of Neurons in Hidden Layer \n The MLP contains '+str(i)+' hidden layer')
    plt.show()
    
    gc.collect()

gc.collect()
```
